##loading libraries and the dataset

```python
import pandas as pd

df = pd.read_csv('/content/Iris.csv')

# Separate features (X) and target variable (y)
X = df.iloc[:, :-1]
y = df.iloc[:, -1]

```

## Create an instance of DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier


classifier = DecisionTreeClassifier()

# Fit method helps to train the data that we all know
classifier.fit(X, y)
```

##For visualization purpose we uses plot_tree and matplotlib dependencies

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Plot the decision tree
fig, ax = plt.subplots(figsize=(12, 8))
plot_tree(classifier, feature_names=X.columns, class_names=df['Species'].unique(), filled=True)
plt.show()
```
